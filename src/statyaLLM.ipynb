# dataset

```python
faculty_dict = {
    0: "IT",
    1: "SISS",
    2: "RIT",
    3: "KIIB",
    4: "CEIMK",
}

reversed_faculty_dict = {
    "IT": 0,
    "SISS": 1,
    "RIT": 2,
    "KIIB": 3,
    "CEIMK": 4,
}
```

```python
texts = new_df.text
labels = [reversed_faculty_dict[x] for x in new_df.faculty]
```

```python
X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.2, random_state=42)
```

```python
class TextDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_length=512):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts.iloc[idx]
        label = self.labels[idx]
        encoding = self.tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )
        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label, dtype=torch.long)
        }
```

# edu

```python
def train(model, train_loader, val_loader, optimizer, loss_fn, epochs=3):
    for epoch in range(epochs):
        model.train()

        for batch in train_loader:
            optimizer.zero_grad()
            input_ids = batch['input_ids']
            attention_mask = batch['attention_mask']
            labels = batch['labels']

            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            logits = outputs.logits

            loss = loss_fn(logits, labels)
            loss.backward()
            optimizer.step()

        print(f'Epoch {epoch+1}/{epochs}')
        print('Evaluating on validation data...')
        evaluate(model, X_test, y_test, tokenizer)
```

```python
from sklearn.metrics import accuracy_score, f1_score
from torch.utils.data import DataLoader

def evaluate(model, X_test, y_test, tokenizer):
    test_dataset = TextDataset(X_test, y_test, tokenizer)
    test_loader = DataLoader(test_dataset, batch_size=2)

    model.eval()
    predictions = []
    true_labels = []

    with torch.no_grad():
        for batch in test_loader:
            input_ids = batch['input_ids']
            attention_mask = batch['attention_mask']
            labels = batch['labels']

            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            logits = outputs.logits
            preds = torch.argmax(logits, dim=1)

            print(f"Logits: {logits}")
            print(f"Predictions: {preds}")
            print(f"True labels: {labels}")

            predictions.extend(preds.cpu().numpy())
            true_labels.extend(labels.cpu().numpy())

    accuracy = accuracy_score(true_labels, predictions)
    f1 = f1_score(true_labels, predictions, average='weighted')
    print(f'Accuracy: {accuracy:.4f}, F1-score: {f1:.4f}')
```

# robert

```python
from transformers import RobertaTokenizer, RobertaForSequenceClassification

tokenizer = RobertaTokenizer.from_pretrained('roberta-base')
model = RobertaForSequenceClassification.from_pretrained('roberta-base', num_labels=5)
```

```python
train_dataset = TextDataset(X_train, y_train, tokenizer)
train_loader = DataLoader(train_dataset, batch_size=2)

val_dataset = TextDataset(X_test, y_test, tokenizer)
val_loader = DataLoader(val_dataset, batch_size=2)
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=5e-5)
```

```python
train(model, train_loader, val_loader, optimizer, nn.CrossEntropyLoss(), epochs=3)
```

```python
evaluate(model, X_test, y_test, tokenizer)
```
